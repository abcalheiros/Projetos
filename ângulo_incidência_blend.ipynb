***Código criado para calcular Ângulos de incidência em modelos no blender***



```python
import bpy
import math
import csv
from mathutils import Vector
from datetime import datetime, timedelta

caminho_arquivo = bpy.path.abspath("//angulo_model1.csv")  # salva na mesma pasta do arquivo .blend

# horário do nascer e pôr do sol (horário solar local)
hora_nascer = datetime(2025, 6, 21, 5, 38)
# cada frame representa 1 minuto

def direcao_sol(sol):
    """Retorna o vetor unitário da direção do Sol"""
    direcao_local = Vector((0, 0, 1))  # mude para (0, 0, -1) se o sentido estiver invertido
    return (sol.matrix_world.to_quaternion() @ direcao_local).normalized()

def normal_painel(painel):
    """Retorna o vetor unitário da normal do painel"""
    normal_local = Vector((0, 0, 1))
    return (painel.matrix_world.to_quaternion() @ normal_local).normalized()

# obtém os objetos
sol = bpy.data.objects["Sol"]
painel = bpy.data.objects["Painel fotovoltaico"]

# cria e escreve o arquivo CSV
with open(caminho_arquivo, "w", newline="", encoding="utf-8") as arquivo_csv:
    escritor = csv.writer(arquivo_csv)
    escritor.writerow(["Frame", "Hora Solar", "Ângulo de Incidência"])

    for frame in range(1, 693 + 1):
        bpy.context.scene.frame_set(frame)
        vetor_sol = direcao_sol(sol)
        vetor_painel = normal_painel(painel)

        produto_escalar = vetor_sol.dot(vetor_painel)
        produto_escalar = max(min(produto_escalar, 1.0), -1.0)
        angulo_incidencia = math.degrees(math.acos(produto_escalar))

        hora_atual = hora_nascer + timedelta(minutes=(frame - 1))
        hora_formatada = hora_atual.strftime("%H:%M")

        escritor.writerow([frame, hora_formatada, round(angulo_incidencia, 3)])

print("salvo em vídeos")
```
